```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('/Northeastern/Fall 2021/ALY6020 Predictive Analytics/Week 1/adult-all.csv', index_col = 0)
```

```python
df
```

```python
df = df.drop(['77516', '2174', '0', '40', '13', 'United-States'], axis = 1)
```

```python
df
```

```python
df.columns = ['Emp Type', 'Education', 'Marital', 'Occupation', 'Household', 'Race', 'Male', 'HighIncome']
```

```python
df = df[(df.Education == 'Bachelors') | (df.Education == 'HS-grad') | (df.Education == 'Masters')]
df = df[(df.Race == 'White') | (df.Race == 'Black') | (df.Race == 'Asian-Pac-Islander')]
```

```python
df
```

```python
df_data = df[['Education', 'Race', 'Male']].copy()
df_labels = df[['HighIncome']].copy()
```

```python
df_data['Male'] = df_data['Male'].replace(['Male'],'1')
df_data['Male'] = df_data['Male'].replace(['Female'], '0')
df_data['Education'] = df_data['Education'].replace(['HS-grad'],'0')
df_data['Education'] = df_data['Education'].replace(['Bachelors'], '1')
df_data['Education'] = df_data['Education'].replace(['Masters'], '2')
df_data['Race'] = df_data['Race'].replace(['White'],'0')
df_data['Race'] = df_data['Race'].replace(['Black'], '1')
df_data['Race'] = df_data['Race'].replace(['Asian-Pac-Islander'], '2')
```

```python
df_data
```

```python
df_labels['HighIncome'] = df_labels['HighIncome'].replace(['<=50K'], '0')
df_labels['HighIncome'] = df_labels['HighIncome'].replace(['>50K'], '1')
```

```python
df_labels
```

```python
from sklearn.model_selection import train_test_split
df_data_train, df_data_test, df_labels_train, df_labels_test = train_test_split(df_data, df_labels, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_data_train)

df_data_train = scaler.transform(df_data_train)
df_data_test = scaler.transform(df_data_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=8)
classifier.fit(df_data_train, df_labels_train)
```

```python
y_pred = classifier.predict(df_data_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(df_labels_test, y_pred))
print(classification_report(df_labels_test, y_pred))
```
