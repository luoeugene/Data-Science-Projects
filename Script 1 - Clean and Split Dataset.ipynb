### Drop unnecessary columns

```python
df = df.drop(columns=['IPC Flag', 'Cust ID', 'ASIN'])
```

### Remove spaces from columns

```python
df.rename(columns = {'ERP Combined Start Fiscal Year': 'ERP_Combined_Start_Fiscal_Year', 'ERP Combined Start Fiscal Week': 'ERP_Combined_Start_Fiscal_Week', 'Product Division': 'Product_Division', 'Style Desc': 'Style_Desc', 'ERP Gender': 'ERP_Gender','ERP Master Gender': 'ERP_Master_Gender', 'ERP Season Cd Opened':'ERP_Season_Cd_Opened', 'US Size': 'US_Size', 'ERP Ooh Qty': 'ERP_Ooh_Qty', 'ERP Ooh': 'ERP_Ooh', 'ERP Cancel Qty': 'ERP_Cancel_Qty', 'ERP Cancel': 'ERP_Cancel', 'ERP Sls Qty': 'ERP_Sls_Qty', 'ERP Sls': 'ERP_Sls'}, inplace = True)
```

### Fill N/A with 0

```python
df['ERP_Ooh_Qty'] = df['ERP_Ooh_Qty'].fillna(0)
df['ERP_Ooh'] = df['ERP_Ooh'].fillna(0)
df['ERP_Sls_Qty'] = df['ERP_Sls_Qty'].fillna(0)
df['ERP_Sls'] = df['ERP_Sls'].fillna(0)
df['ERP_Cancel_Qty'] = df['ERP_Cancel_Qty'].fillna(0)
df['ERP_Cancel'] = df['ERP_Cancel'].fillna(0)
```

### Merge the color column to the main data set

```python
new.columns
```

```python
new.drop(['Article[Article]', 'Article[Color Desc]',
       'Article[ERP Gender]', 'Article[ERP Master Gender]',
       'Article[ERP Season Cd Opened]',
       'Article[Product Division]', 'Article[Reporting Business Unit]',
       'Article[Style Desc]', 'Article[Style Family]', 'Article[Style Num]',
       'SKU[Height (in)]', 'SKU[Length (in)]', 'SKU[US Size]',
       'SKU[Weight (g)]', 'SKU[Width (in)]', 'ERP_Master_Gender'], axis=1, inplace=True)
```

### Rename column color

### Add Total Demand Column

```python
new['Total_Demand_Qty'] = new['Cancelled_Qty'] + new['Shipped_Qty'] + new['On_Hand_Qty']
new['Total_Demand_Value'] = new['On_Hand_Value'] + new['Cancelled_Value'] + new['Shipped_Value']
new.head()
```

### Make Price Column

```python
new['Price'] = new['Total_Demand_Value']/new['Total_Demand_Qty']
new.head()
```
