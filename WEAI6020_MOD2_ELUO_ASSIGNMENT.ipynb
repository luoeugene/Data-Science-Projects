```python
# Create a data set of 1000 random numbers starting at 1/1/2022 
# Use the pandas cumsum method to calculate the cumulative sum
# Examine the data and make sure that you understand what is going on
ts_1 = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2022', periods=1000))
ts_1.head(10)
ts_2 = ts_1.cumsum()
ts_2.head(10)
ts_2.plot()
plt.title('1000 Random Number Series')
plt.xlabel('Date Range')
```

```python
# Generate 1000 data points as in the previous cell - 
# give it a different object name 
# and plot the series

ts_2 = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2022', periods=1000))
ts_2.head(10)
ts_2 = ts_2.cumsum()
ts_2.head(10)
ts_2.plot()
plt.title('1000 Random Number Series')
plt.xlabel('Date Range')
```

```python
# Create a data frame with 5 series of 1000 data points
# Uses the ts_1 index created in the previous cells
# Plot the time series using plot. 

df_1 = pd.DataFrame(np.random.randn(1000,5), index=ts_1.index, columns=list('ABCDE'))
df_1.head(10)
df_1 = df_1.cumsum()
df_1.head(10)
df_1.plot()

plt.title('1000 Random Number 5 Series')
plt.xlabel('Date Range')
```

```python
# Plot the data points in the above data frame
plt.figure(); 

df_1.plot()
plt.title('1000 Random Number 5 Series')
plt.xlabel('Date Range')
```

```python
# The purpose of using plt.figure() is to create a figure object.
# The whole figure is regarded as the figure object. 
# It is necessary to explicitly use plt.figure() 
# when we want to tweak the size of the figure and 
# when we want to add multiple Axes objects in a single figure.

plt.figure();

# .iloc[] is primarily integer position based - from 0 to length-1 of the 
# axis; Here we use the slice object - to create a 5 by 5 data frame
# Take a slice of 10 data points and create bar plots 
# .iloc[] is primarily integer position based - from 0 to length-1 of the 
# axis; Use the slice object - to create a 10 by 10 data frame
# Plot the bar plot and use plt.axhline(0, color='r') to 
# draw a red hline at 'y' = 0 that spans the xrange:

df_1.iloc[0:5,0:5].plot.bar();

df_1.iloc[0:10,0:10].plot.bar();

plt.axhline(0, color="r");

plt.title('Bar Plot Example')
plt.xlabel('Variable')
```

```python
# Create a data frame with three variables of 1000 observations. The first
# should be a set of random numbers with an increment of +2, 
# the second should be purely random numbers and the third a set of 
# random numbers with an increment of -2.
# Display the first 5 observation using head, the shape of the data frame 
# and summary statistics using describe()

df_3 = pd.DataFrame(
    {
        'a': np.random.randn(1000) + 2,
        'b': np.random.randn(1000),
        'c': np.random.randn(1000) - 2,
    },
    columns=['a','b','c'],
)

df_3.head()

df_3.shape

df_3.describe()
```

```python
# Histograms can be drawn using the DataFrame.plot.hist() and 
# Series.plot.hist() methods. 
# Plot the histogram of the above data frame created in the previous cell


plt.figure();

df_3.plot.hist(alpha=0.5);

plt.title('Histogram')
plt.xlabel('Range')
```

```python
# A histogram can be stacked using stacked=True. 
# Draw a stacked histogram of the above dataframe
# Bin size can be changed using the bins keyword.

df_3.plot.hist(stacked=True, bins=20);
plt.title('Histogram')
plt.xlabel('Range')
```

```python
# Boxplot can be drawn calling Series.plot.box() and DataFrame.plot.box(),
# or DataFrame.boxplot() to visualize the distribution of 
# values within each column.
# Draw a boxplot representing 10 trials of 10 observations of a 
# uniform random variable on [0,1).

df_4 = pd.DataFrame(np.random.rand(10, 10), columns=list('ABCDEFGHIJ'))

df_4.plot.box();

plt.title('Boxplot')
plt.xlabel('Variable')
```

```python
# Boxplot can be colorized by passing color keyword. 
# You can pass a dict whose keys are boxes, whiskers, medians and caps. 
# If some keys are missing in the dict, 
# default colors are used for the corresponding artists. 
# Also, boxplot has sym keyword to specify fliers style.

# When you pass other type of arguments via color keyword, 
# it will be directly passed to matplotlib for all the boxes, 
# whiskers, medians and caps colorization.

# The colors are applied to every boxes to be drawn. 
# If you want more complicated colorization, 
# you can get each drawn by passing return_type.
# Colorize the box plots: boxes: green, whiskers: orange, 
# medians: blue and caps as red.

color = {
    'boxes': 'green',
    'whiskers': 'orange',
    'medians': 'blue',
    'caps': 'red',
}

df_4.plot.box(color=color, sym='r+');

plt.title('Boxplot')
plt.xlabel('Variable')
```

```python
# Scatter Plots - create a data frame with 100 random numbers and 
# 5 variables. Use head and shape

df_5 = pd.DataFrame(np.random.rand(100,5), columns=list('ABCDE'))

df_5.head()

df_5.shape
```

```python
# Create a scatter plot of any two variables from the dataframe in the 
# above cell

df_5.plot.scatter(x='A', y='B');

plt.title('Scatterplot')
plt.xlabel('Range')
```

```python
# Plot multiple column groups in a single axes, 
# Repeat plot method specifying target axis
# Create axes for the first using blue color and label the groups appropriately

ax = df_5.plot.scatter(x='A', y='B', color='blue', label='Group 1')

df_5.plot.scatter(x='C', y='D', color='green', label='Group 2', ax=ax)

plt.title('Scatterplot')
plt.xlabel('Range')
```

```python
# Create a bubble chart using a scatter plot for two variables and 
# a column of the data frame as an index of size. 
# Your result should show a scatter plot with the third variable as a 
# proxy for size

df_5.plot.scatter(x='A', y='B', s=df_5['C'] * 250)

plt.title('Scatterplot')
plt.xlabel('Range')
```

```python
# Plotting Tools
# These functions can be imported from pandas.plotting and take a 
# Series or DataFrame as an argument.

# Scatter Matrix Plot
# You can create a scatter plot matrix using the scatter_matrix 
# method in pandas.plotting:
# alpha is the amount of transparency applied
# diagonal - Pick between ‘kde’ and ‘hist’ for either Kernel Density 
# Estimation or Histogram plot in the diagonal.
# In statistics, kernel density estimation (KDE) is a non-parametric way 
# to estimate the probability density function (PDF) of a random variable.
# Vary alpha and observe how transparency works
# Use kde in the first plot and hist in the second
from pandas.plotting import scatter_matrix

# First generate a dataframe of 1000 observatins with 5 variables
# Use pd.plotting to create the plots as shown in class

df_6 = pd.DataFrame(np.random.rand(1000,5), columns = list('ABCDE'))

scatter_matrix(df_6, alpha=0.5, figsize=(7,7), diagonal='kde')
scatter_matrix(df_6, alpha=0.4, figsize=(7,7), diagonal='hist')
```
